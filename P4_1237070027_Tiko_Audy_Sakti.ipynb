```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("upload 4 gambar")
uploaded = files.upload()
images = {}
for filename in uploaded.keys():
    # Mengubah data bytes dari file yang diupload
    file_bytes = np.frombuffer(uploaded[filename], np.uint8)
    img = cv2.imdecode(file_bytes, cv2.IMREAD_GRAYSCALE)
    images[filename] = img

if len(images) > 0:
    print("\nHASIL SOAL 1: CITRA ASLI & HISTOGRAMNYA")
    plt.figure(figsize=(12, 3 * len(images)))

    for i, (filename, img) in enumerate(images.items()):
        # Menampilkan Gambar Asli
        plt.subplot(len(images), 2, i*2 + 1)
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
        plt.title(f'Citra Asli: {filename}')
        plt.axis('off')

        # Menampilkan Histogram Asli
        plt.subplot(len(images), 2, i*2 + 2)
        plt.hist(img.ravel(), 256, [0,256], color='red', alpha=0.7)
        plt.title(f'Histogram: {filename}')
        plt.xlim([0, 256])

    plt.tight_layout()
    plt.show()

    print("\n HASIL SOAL 2: NORMALISASI (HISTOGRAM EQUALIZATION)")
    plt.figure(figsize=(12, 3 * len(images)))

    for i, (filename, img) in enumerate(images.items()):
        # Melakukan Normalisasi / Ekualisasi Histogram
        img_eq = cv2.equalizeHist(img)

        # Menampilkan Gambar Hasil Normalisasi
        plt.subplot(len(images), 2, i*2 + 1)
        plt.imshow(img_eq, cmap='gray', vmin=0, vmax=255)
        plt.title(f'Citra Normalisasi: {filename}')
        plt.axis('off')

        # Menampilkan Histogram Hasil Normalisasi
        plt.subplot(len(images), 2, i*2 + 2)
        plt.hist(img_eq.ravel(), 256, [0,256], color='blue', alpha=0.7)
        plt.title(f'Histogram Normalisasi: {filename}')
        plt.xlim([0, 256])

    plt.tight_layout()
    plt.show()
```

soal 3 power law

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage.metrics import mean_squared_error, peak_signal_noise_ratio, structural_similarity
from google.colab import files

print("Upload gambar")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img_original = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Fungsi buat Gamma Correction
def apply_gamma(img, gamma):
    # Normalisasi ke [0, 1]
    img_norm = img / 255.0
    # Rumus Power Law: s = r^gamma
    gamma_corrected = np.power(img_norm, gamma)
    # Balikin ke [0, 255]
    return (gamma_corrected * 255).astype(np.uint8)

# Daftar Gamma sesuai tugas
gammas_display = [0.3, 0.7, 1.0, 1.5, 2.5]
gammas_metrics = [0.1, 0.5, 1.2, 2.2]

# visualiasai tugas 1 dan 2
plt.figure(figsize=(15, 5))
for i, g in enumerate(gammas_display):
    res = apply_gamma(img_original, g)
    plt.subplot(1, len(gammas_display), i+1)
    plt.imshow(res, cmap='gray', vmin=0, vmax=255)
    plt.title(f'Gamma = {g}')
    plt.axis('off')
plt.show()

# tugas 5: tabel metrik
results = []
for g in gammas_metrics:
    res = apply_gamma(img_original, g)
    mse = mean_squared_error(img_original, res)
    psnr = peak_signal_noise_ratio(img_original, res, data_range=255)
    ssim = structural_similarity(img_original, res, data_range=255)
    results.append({'Gamma': g, 'MSE': mse, 'PSNR': psnr, 'SSIM': ssim})

df = pd.DataFrame(results)
print("tabel perbandingan metriks")
print(df.to_string(index=False))
```

soal 4. contrast scretching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("Upload gambar yang kontrasnya rendah")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Membaca gambar dalam grayscale
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Proses Contrast Stretching
r_min = np.min(img)
r_max = np.max(img)

# Menghindari pembagian dengan nol jika gambar isinya satu warna semua
if r_max == r_min:
    img_stretched = img
else:
    img_stretched = ((img - r_min) * (255 / (r_max - r_min))).astype(np.uint8)

# Menampilkan hasil
plt.figure(figsize=(12, 8))

# Sebelum
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Citra Asli (Low Contrast)')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), 256, [0,256], color='red', alpha=0.7)
plt.title('Histogram Sebelum Stretching')

# Sesudah
plt.subplot(2, 2, 3)
plt.imshow(img_stretched, cmap='gray', vmin=0, vmax=255)
plt.title('Citra Hasil Contrast Stretching')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(img_stretched.ravel(), 256, [0,256], color='blue', alpha=0.7)
plt.title('Histogram Sesudah Stretching')

plt.tight_layout()
plt.show()
```

soal 6. CLAHE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from skimage.metrics import mean_squared_error, peak_signal_noise_ratio, structural_similarity

print("upload gambar")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# proses Histogram Equalization
img_he = cv2.equalizeHist(img)

# proses CLAHE
clip_limit = 2.0
grid_size = (8, 8)
clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=grid_size)
img_clahe = clahe.apply(img)

# fungsi hitung metrik
def get_metrics(img_ref, img_test):
    mse = mean_squared_error(img_ref, img_test)
    psnr = peak_signal_noise_ratio(img_ref, img_test, data_range=255)
    ssim = structural_similarity(img_ref, img_test, data_range=255)
    return mse, psnr, ssim

mse_he, psnr_he, ssim_he = get_metrics(img, img_he)
mse_clahe, psnr_clahe, ssim_clahe = get_metrics(img, img_clahe)

# visualisasi
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img, cmap='gray'); plt.title('Asli'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(img_he, cmap='gray'); plt.title('Histogram Equalization'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(img_clahe, cmap='gray'); plt.title('CLAHE'); plt.axis('off')
plt.show()

# 6. Output Metrik (Silakan copy-paste angka ini untuk saya analisis)
print(f"\n=== METRIK HE ===")
print(f"MSE  : {mse_he:.4f}")
print(f"PSNR : {psnr_he:.4f}")
print(f"SSIM : {ssim_he:.4f}")

print(f"\n=== METRIK CLAHE ===")
print(f"MSE  : {mse_clahe:.4f}")
print(f"PSNR : {psnr_clahe:.4f}")
print(f"SSIM : {ssim_clahe:.4f}")
```
