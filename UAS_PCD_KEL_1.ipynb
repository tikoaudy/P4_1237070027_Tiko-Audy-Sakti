```python
import cv2
import numpy as np
import gradio as gr

# 1. Fungsi Utama Segmentasi Thresholding
def process_image(input_image):
    if input_image is None:
        return None, None, None, None, "⚠️ wilujeng unggah gambar."

    # Konversi input ke Grayscale sebagai syarat wajib Thresholding
    gray_image = cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY)

    # Metode 1: Global Thresholding
    _, global_thresh = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    # Metode 2: Adaptive Thresholding
    adaptive_thresh = cv2.adaptiveThreshold(
        gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )

    # Metode 3: Otsu's Thresholding
    otsu_val, otsu_thresh = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Tabel Evaluasi & Analisis Studi Kasus
    metrics_text = f"""
    ### 📊 Hasil Analisis: Ekstraksi Objek pada Kondisi Berbayang
    *Tujuan: Memisahkan teks/objek dari latar belakang yang memiliki iluminasi tidak merata.*

    | Metode Thresholding | Cara Kerja Algoritma | Evaluasi Kinerja Terhadap Bayangan |
    | :--- | :--- | :--- |
    | **1. Global Thresholding** | Menggunakan satu nilai batas (127) untuk seluruh area foto. | **Buruk.** Area yang tertutup bayangan akan dianggap hitam pekat, membuat objek/teks hilang atau menyatu dengan bayangan. |
    | **2. Adaptive Thresholding** | Menghitung rata-rata cahaya di blok-blok kecil area sekitar piksel. | **Sangat Baik.** Mampu mengabaikan bayangan gelap karena sistem mengevaluasi kontras secara lokal. Sangat cocok untuk *scanner* dokumen. |
    | **3. Otsu's Thresholding** | Mesin mencari nilai batas otomatis **(Nilai Otsu = {otsu_val})** berdasarkan histogram. | **Kurang Stabil.** Sangat cerdas pada gambar normal, namun sering gagal memisahkan objek jika bayangannya terlalu pekat. |

    *💡 **Kesimpulan:** Pada kondisi pencahayaan yang berbayang atau tidak merata, **Adaptive Thresholding** terbukti menjadi metode paling andal untuk mengekstrak objek tanpa merusak detail aslinya.*
    """

    return gray_image, global_thresh, adaptive_thresh, otsu_thresh, metrics_text

# 2. Rancang Bangun Antarmuka (UI) dengan Gradio
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    # Judul
    gr.Markdown("# 🔲 Penerapan Metode Thresholding pada Sistem Segmentasi Citra Berbasis Web")
    gr.Markdown("**Tujuan:** Mengatasi Kendala Pencahayaan Berbayang Menggunakan Gradio dan Google Colab.")

    with gr.Row():
        with gr.Column():
            img_input = gr.Image(label="1. Upload Foto Di Sini (Sangat disarankan: Foto dokumen/struk yang terhalang bayangan)")
            btn_proses = gr.Button("Mulai Segmentasi Biner", variant="primary")
        with gr.Column():
            metrics_output = gr.Markdown() #analisis performa

    with gr.Row():
        img_gray = gr.Image(label="Tahap Pra-pemrosesan: Citra Grayscale", interactive=False)
        img_global = gr.Image(label="Hasil 1: Global Thresholding (Rentan Bayangan)", interactive=False)

    with gr.Row():
        img_adaptive = gr.Image(label="Hasil 2: Adaptive Thresholding (Paling Tahan Bayangan)", interactive=False)
        img_otsu = gr.Image(label="Hasil 3: Otsu's Thresholding", interactive=False)

    # Menghubungkan tombol klik dengan fungsi Python
    btn_proses.click(
        fn=process_image,
        inputs=img_input,
        outputs=[img_gray, img_global, img_adaptive, img_otsu, metrics_output]
    )

# 3. running web gradio
demo.launch(share=True, debug=True)
```
